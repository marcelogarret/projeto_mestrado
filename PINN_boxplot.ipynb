```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregar os dados do arquivo Excel
df=pd.read_excel('/content/drive/MyDrive/UNISIM/results_pinn_mae.xlsx')
df.head()
```

```python
# Configurações do estilo do gráfico
sns.set(style="whitegrid")

# Criando o boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette="Set1")
plt.title("Boxplot do MAE para diferentes tipos de redes neurais", fontsize=16)
# plt.xlabel("Tipo de Rede Neural", fontsize=14)
plt.ylabel("MAE (Mean Absolute Error)", fontsize=14)
plt.show()
```

```python
def boxplot_sorted(df, rot=90, figsize=(12,6), fontsize=20):
    meds = df.median().sort_values(ascending=False)
    axes = df[meds.index].boxplot(figsize=figsize, rot=rot, fontsize=fontsize,
                                   boxprops=dict(linewidth=4, color='cornflowerblue'),
                                   whiskerprops=dict(linewidth=4, color='cornflowerblue'),
                                   medianprops=dict(linewidth=4, color='firebrick'),
                                   capprops=dict(linewidth=4, color='cornflowerblue'),
                                   flierprops=dict(marker='o', markerfacecolor='dimgray',
                                        markersize=12, markeredgecolor='black'),
                                   return_type="axes")

    axes.set_title("Boxplot do MAE para diferentes tipos de redes neurais", fontsize=fontsize)
    # axes.set_xlabel("Tipo de Rede Neural", fontsize=fontsize)
    axes.set_ylabel("MAE (Mean Absolute Error)", fontsize=fontsize)

boxplot_sorted(df, rot=90, figsize=(10,5), fontsize=15)
```
